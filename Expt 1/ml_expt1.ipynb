```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = "homeprices.csv" 
df = pd.read_csv(file_path)
```

```python
print("üìÑ DATASET STRUCTURE")
print("-" * 50)
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
print("Column Names:", list(df.columns))
```

```python
print("\nüîç COLUMN DATA TYPES & MISSING VALUES")
print("-" * 50)
df.info()
```

```python
print("\nüìä NUMERICAL STATISTICS")
print("-" * 50)
print(df.describe())
```

```python
print("\nüìà FREQUENCY COUNTS FOR CATEGORICAL COLUMNS")
print("-" * 50)
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    print(f"\n{col.upper()} value counts:")
    print(df[col].value_counts(dropna=False))
```

```python
print("\nüîó CORRELATION MATRIX")
print("-" * 50)
print(df.corr(numeric_only=True))
```

```python
sns.set(style="whitegrid")
```

```python
df.hist(bins=20, figsize=(12, 6), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Columns", fontsize=16)
plt.show()
```

```python
df.select_dtypes(include=['number']).plot(kind='box', subplots=True, layout=(2, 3), figsize=(12, 6), sharex=False, sharey=False)
plt.suptitle("Boxplots for Numeric Columns", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
